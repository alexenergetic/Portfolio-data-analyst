### Просмотр распределения данных

Посмотрим распределение данных по столбцам и выявим возможные проблемы.

Создадим функции для автоматизации построения графиков.

```python
# создадим список столбцов для построения по ним графика гистограммы
columns_for_hist = ['score', 'age', 'objects', 'balance', 'estimated_salary']

# создадим списки для подписей графиков и подписей оси х
columns_hist_titles = ['баллов кредитного скоринга', 'возраста', 'количества объектов в собственности', \
                       'баланса счета', 'предполагаемой зарплаты']
columns_hist_xlabels = ['Количество баллов', 'Возраст', 'Количество объектов', \
                        'Баланс счета', 'Размер предполагаемой зарплаты']
```

```python
# напишем функцию для автоматизации построения гистограмм 
def create_hist(list_columns, list_titles, list_xlabels):
    '''Функция принимает 1 аргумент на вход список со столбцами, по которым построит гистограмму.
       2 аргумент - список для подписи графиков, 3 аргумент - список для подписи оси х.
    '''
    
    # задаём размер сетки
    plt.figure(figsize=(30, 80))  
    # задаём счетчик для номера ячейки в subplot
    i = 1
    # задаём счетчик для подписи графика и оси х
    j = 0
    
    for col in list_columns:
        df[col].hist(figsize=(15,20), ax=plt.subplot(3, 2, i))
        plt.title('Гистограмма {} клиента'.format(list_titles[j]))
        plt.xlabel(list_xlabels[j])
        plt.ylabel('Частота')
        # увеличиваем счетчики
        i+= 1
        j+= 1
    plt.show()
```

```python
# вызовем функцию и построим гистограммы
create_hist(columns_for_hist, columns_hist_titles, columns_hist_xlabels)
```

Распределение по баллам кредитного скоринга близкое к нормальному, но видны выбросы по максимальным количеством баллов кредитного скоринга. Такие клиенты могут существовать. Большинство клиентов имеет от 550 до 750 баллов. 

Распределение по возрасту скошено вправо. Большая часть клиентов до 50 лет. Есть клиенты и 90 летние. Такие клиенты также могут существовать.

По распределению количества объектов в собственности видно, что владение от 1 до 9 объектов равномерно распределено среди клиентов. При этом заметно, что мало клиентов без собственности. Также видно, есть группа клиентов с 10 объектами собственности.

Баланс счета клиентов имеет нормальное распределение. Большинство имеет баланс в диапазоне 80 - 150 тысяч.

Предполагаемая зарплата клиента равномерно распределена среди всех клиентов. Что немного странно, т.к. ожидается кривая распределения. Возможно есть проблемы с алгоритмом определения зарплаты у клиента.

```python
# создадим список столбцов для построения по ним cтолбчатой диаграммы
columns_for_bar = ['city', 'gender', 'products', 'credit_card', 'loyalty', 'churn']

# создадим списки для подписей графиков и подписей оси х
columns_bar_titles = ['по городам', 'пола', 'количества банковских продуктов у', \
                      'банковской карты у', 'лояльности', 'оттока']
columns_bar_xlabels = ['Местожительство клиента', 'Пол клиентов', 'Количество продуктов',\
                       'Наличие банковской карты', 'Лояльность клиентов', 'Клиент ушел или нет']
```

```python
# напишем функцию для автоматизации построения cтолбчатой диаграммы 
def create_bar(list_columns, list_titles, list_xlabels):
    '''Функция принимает 1 аргумент на вход список со столбцами, по которым построит cтолбчатую диаграмму.
       2 аргумент - список для подписи графиков, 3 аргумент - список для подписи оси х.
    '''
    
    # задаём размер сетки
    plt.figure(figsize=(30, 80))  
    # задаём счетчик для номера ячейки в subplot
    i = 1
    # задаём счетчик для подписи графика и оси х
    j = 0
    
    for col in list_columns:
        df.groupby(col)[col].count().plot(kind='bar', figsize=(15,20), ax=plt.subplot(3, 2, i))
        plt.title('Распределение {} клиентов'.format(list_titles[j]))
        plt.xlabel(list_xlabels[j])
        plt.ylabel('Количество клиентов')
        plt.xticks(rotation=0)
        # увеличиваем счетчики
        i+= 1
        j+= 1
    plt.show()
```

```python
# вызовем функцию и построим гистограммы
create_bar(columns_for_bar, columns_bar_titles, columns_bar_xlabels)
```

Число лояльных клиентов немного больше, чем число нелояльных клиентов. Перевес небольшой.

### Категоризация клиентов и добавление новых столбцов

Сделаем категоризацию по возрасту, по балансу счета и по предполагаемой зарплате клиента, а также по баллам кредитного скоринга.

Методом describe() мы узнали, что возраст клиентов от 18 до 92 лет. Сделаем категоризацию диапазонами по 10 лет, кроме возрастных людей от 60 лет. Их включим в 1 категорию.

```python
# создадим функцию категоризации
def age_category(age):
    '''
    Функция принимает на вход значения возраста клиента и присваивает категорию согласно шкале.
    
    Применим конструкцию try - except, чтобы пропустить возможные ошибки.
    '''
    try:
        if age >= 60:
            return 'от 60 лет'
        elif 50 <= age < 60:
            return '50 - 59 лет'
        elif 40 <= age < 50:
            return '40 - 49 лет'
        elif 30 <= age < 40:
            return '30 - 39 лет'   
        elif 18 <= age < 30:
            return '18 - 29 лет'
    except:
        pass
```

```python
# применим функцию категоризации, сделаем новый столбец с категорией по возрасту
df['age_category'] = df['age'].apply(age_category)
```

Методом describe() мы узнали, что баланс клиентов от 3768 до 250898 лет. Сделаем категоризацию баланса диапазонами по 50000.

```python
# создадим функцию категоризации
def balance_category(balance):
    '''
    Функция принимает на вход значения баланса клиента и присваивает категорию согласно шкале.
    
    Если баланс содержит пропуск, ставится значение 'неизвестно'.
    '''
    
    if balance >= 200000:
        return 'свыше 200 тыс'
    elif 150000 <= balance < 200000:
        return 'от 150 до 200 тыс'
    elif 100000 <= balance < 150000:
        return 'от 100 до 150 тыс'
    elif 50000 <= balance < 100000:
        return 'от 50 до 100 тыс'   
    elif balance < 50000:
        return 'до 50 тыс'
        
    return 'неизвестно'
```

```python
# применим функцию категоризации, сделаем новый столбец с категорией по возрасту
df['balance_category'] = df['balance'].apply(balance_category)
```

Создадим функцию для автоматизации группировки данных по столбцам и построения графиков для поиска закономерностей пропусков в столбце balance.

```python
# создадим список столбцов для построения по ним столбчатых графиков 
columns_for_barplot = ['score_category', 'city', 'gender', 'age_category', 'objects', \
                       'products', 'credit_card', 'loyalty', 'salary_category', 'churn']

# создадим списки для подписей графиков и подписей легенды
columns_barplot_titles = ['Баллов кредитного скоринга', 'города', 'пола', \
                      'возраста', 'объектов в собственности', 'продуктов у', \
                         'наличия кредитной карты', 'лояльности', 'зарплаты', 'оттока']
columns_barplot_legend = ['Баллы кредитного скоринга', 'Город', 'Пол', \
                       'Возраст', 'Количество объектов', 'Количество продуктов', \
                         'Наличие кредитной карты', 'Лояльность', 'Зарплата', 'Отток']
```

```python
# напишем функцию для автоматизации построения cтолбчатой диаграммы 
def create_barplot(list_columns, list_titles, list_legend):
    '''Функция принимает 1 аргумент на вход список со столбцами. 
       Делает группировку по наличию или отсутсвию пропусков в столбце баланс вместе с столбцом из списка.
       Далее по группировке строится barplot.
       2 аргумент - список для подписи графиков, 3 аргумент - список для подписи легенды.
    '''
    
    # задаём размер сетки
    plt.figure(figsize=(30, 150))  
    # задаём счетчик для номера ячейки в subplot
    i = 1
    # задаём счетчик для подписи графика и оси х
    j = 0
    
    for col in list_columns:
        # сделаем группировку по наличию или отсутсвию пропусков в столбце баланс вместе с баллами кредитного скоринга
        balance_group = df.groupby([df['balance'].isnull(), col])['user_id'].count().reset_index()
        # найдем процент записей в разбивке от общего числа в датафрейме
        balance_group['percent'] = balance_group['user_id'] / balance_group['user_id'].sum() * 100
        
        # построим столбчатый графика
        #зададим размер графика
        sns.set(rc={'figure.figsize':(15,15)})
        sns.barplot(x='balance', y='percent', data=balance_group, hue=col, \
                     ax=plt.subplot(10, 2, i))
        plt.title('Распределение пропусков в столбце balance в разрезе {} клиента'.format(list_titles[j]))
        plt.xlabel('Наличие пропусков в столбце balance')
        plt.ylabel('Количество записей от общего числа в датафрейме в %')
        plt.legend(title=format(list_legend[j]))
        plt.xticks([0,1],['Да', 'Нет'])
        # увеличиваем счетчики
        i+= 1
        j+= 1
        
    plt.show()
```

```python
# вызовем функцию и построим гистограммы
create_barplot(columns_for_barplot, columns_barplot_titles, columns_barplot_legend)
```

Итак, в столбце `balance` пропусков нет у клиентов из Ростова Великого. Пропуски встречаются только у клиентов из Рыбинска и Ярославля.

В разрезе по количеству продуктов у клиента видно, что много с пропусков в записях у клиентов с 2 продуктами.

По остальным параметрам отклонений нет в распределении между строками с пропусками и без пропусков в столбце баланса клиента.

Какую-то законоверность появления пропусков не получается выделить. Предположим, что произошел сбой хранения или выгрузки данных по клиентам из Рыбинска и Ярославля.

## Проведение статистического исследования

### Постановка статистических гипотез

**Задача №1:** Проверьте гипотезу различия дохода между теми клиентами, которые ушли и теми, которые остались.

Теперь составим нулевую и альтернативную гипотезу: 

*H0: Средний доход клиентов, которые остались = Среднему доходу клиентов, которые ушли*

*H1: Средний доход клиентов, которые остались ≠ Среднему доходу клиентов, которые ушли*

Чтобы проверить эти гипотезы, сравним выборки на равенство их средних. Эти выборки не зависимы друг от друга, поэтому используем t-тест.

После проведения теста, мы проверим равенство средних выборок. Если средние выборок будут равны, мы примем нулевую гипотезу (H0), а если средние выборок будет не равны - мы будем отвергать нулевую гипотезу (H0) и принимать альтернативную (H1).

**Задача №2:** Сформулируйте и проверьте статистическую гипотезу относительно представленных данных.

Проверим различие лояльности между теми клиентами, которые ушли и теми, которые остались.

Составим нулевую и альтернативную гипотезу: 

*H0: Доля лояльных клиентов среди клиентов, которые остались =  Доле лояльных клиентов среди клиентов, которые ушли*

*H1: Доля лояльных клиентов среди клиентов, которые остались ≠ Доле лояльных клиентов среди клиентов, которые ушли*

Данные по лояльности клиентов - бинарные. Чтобы проверить эти гипотезы, сравним выборки на равенство долей. Для этого используем z-тест.

После проведения теста, мы проверим разницу пропорций. Если пропорции выборок будут схожи, мы примем нулевую гипотезу (H0), а если спропорции выборок не будут схожи - мы будем отвергать нулевую гипотезу (H0) и принимать альтернативную (H1).

### Проведение статистических тестов и проверка статистических гипотез

#### Проверка статистической гипотезы №1

```python
# сделаем срезы данных по клиентам ушедшим и оставшимся
revenue_stay = df[df['churn'] == 0]
revenue_leave = df[df['churn'] == 1]
```

До статистического t-теста нало узнать дисперсию у выборок. Для этого узнаем длину срезов.

```python
# узнаем длины срезов
display(len(revenue_stay['estimated_salary']))
display(len(revenue_leave['estimated_salary']))
```

Длины различны у срезов, дисперсию будем считать разной. Выборки большие, поэтому уровень значимости поставим 5%.

```python
# постамим уровень значимости для теста
alpha = 0.05

# в срезе есть пропуски, поставим параметр nan_policy для обработки пропуски
# параметр equal_var задаем для различной дисперсии
test_revenue = st.ttest_ind(revenue_stay['estimated_salary'], revenue_leave['estimated_salary'], equal_var = False, nan_policy='omit')

# выводим результат теста на экран
if test_revenue.pvalue < alpha:
    print('Мы отвергаем нулевую гипотезу при значении p-value:', round(test_revenue.pvalue, 3))
else:
    print('У нас не получилось отвергнуть нулевую гипотезу при значении p-value:',  round(test_revenue.pvalue, 3))
```

Видим, что значение p-value проходит уровень критической значимости 5%. Значение p-value больше уровня критической значимости. Поэтому мы не можем отвергнуть нулевую гипотезу.

Итак, мы принимаем нулевую гипотезу, что *Средний доход клиентов, которые остались = Среднему доходу клиентов, которые ушли*.

#### Проверка статистической гипотезы №2

Так как выборки большие, поэтому в этом тесте уровень значимости оставим 5%.

```python
# сделаем срезы данных по клиентам ушедшим и оставшимся
loyalty_stay = df[df['churn'] == 0]['loyalty']
loyalty_leave = df[df['churn'] == 1]['loyalty']
```

```python
# найдем лояльных клиентов в 1 и 2 группе
loyal1 = loyalty_stay.sum()
loyal2 = loyalty_leave.sum()

# найдем общее число клиентов в 1 и 2 группе
cnt1 = loyalty_stay.count()
cnt2 = loyalty_leave.count()

# найдем пропорцию по группе 1
p1 = loyal1 / cnt1

# найдем пропорцию по группе 2
p2 = loyal2 / cnt2

# посчитаем общую пропорция по скомбинированной выборке
p_combined = (p1 + p2) / (cnt1 + cnt2)

# определим разницу пропорций в выборках
diff = p1 - p2

# найдем статистику стандартного отклонения по стандартному нормальному распределению
z_value = diff / mth.sqrt(p_combined * (1 - p_combined) * (1 / cnt1 + 1 / cnt2))

# зададим стандартное нормальное распределение
distr = st.norm(0, 1)

# посчитаем значение p-value
p_value = (1 - distr.cdf(abs(z_value))) * 2

# выводим результат теста на экран
if p_value < alpha:
    print('Есть значимая разница между долями. Мы отвергаем нулевую гипотезу при значении p-value:', round(p_value, 2))
else:
    print('Нет значимых разниц между долями. У нас не получилось отвергнуть нулевую гипотезу при значении p-value:', round(p_value, 2))
```

Значение p-value меньше уровня критической значимости в 5%. Поэтому мы отвергаем нулевую гипотезу.

Итак, мы принимаем альтернативную гипотезу, что *Доля лояльных клиентов среди клиентов, которые остались ≠ Доле лояльных клиентов среди клиентов, которые ушли*.
