### Примените фреймворк ICE для приоритизации гипотез. Отсортируйте их по убыванию приоритета.

Шкала приоритизации гипотез ICE рассчитывается по значениям impact, confidence, effort.

```python
# создадим в датафрейме столбец с расчетом шкалы ICE
data['ICE'] = round((data['impact'] * data['confidence']) / data['efforts'], 2)
# пронумеруем гипотезы
data['number'] = data.index + 1
```

```python
# выведим список гипотез c рассчитанной шкалой ICE по убыванию
data[['number', 'hypothesis', 'ICE']].sort_values(by='ICE', ascending=False)
```

**Вывод:** По фреймворку ICE наиболее приоритетны 3 гипотезы под номерами:
   - №9 (Запустить акцию, дающую скидку на товар в день рождения),
   - №1 (Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей),
   - №8 (Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок).

### Примените фреймворк RICE для приоритизации гипотез. Отсортируйте их по убыванию приоритета.

Шкала приоритизации гипотез RICE рассчитывается по значениям reach, impact, confidence и effort.

```python
# создадим в датафрейме столбец с расчетом шкалы RICE
data['RICE'] = (data['reach'] * data['impact'] * data['confidence']) / data['efforts']
```

```python
# выведим список гипотез c рассчитанной шкалой RICE по убыванию
data[['number', 'hypothesis', 'RICE']].sort_values(by='RICE', ascending=False)
```

**Вывод:** По фреймворку RICE наиболее приоритетны 4 гипотезы под номерами:
   - №8 (Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок),
   - №3 (Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказ),
   - №1 (Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей),
   - №7 (Показать на главной странице баннеры с актуальными акциями и распродажами, чтобы увеличить конверсию).
   
  Причем гипотезы №1 и №7 занимают одно место, т.к. набрали одинаковое количество баллов.

### Укажите, как изменилась приоритизация гипотез при применении RICE вместо ICE. Объясните, почему так произошло.

Гипотеза №9 (Запустить акцию, дающую скидку на товар в день рождения) имеет малых охват поэтому ушла из Топ-3 гипотез.

Таким образом оба фрейморка показывают актуальные гипотезы. Отличия только в том, что RICE учитывает охват аудитории.

```python
# создадим список уникальных пар дат и групп теста
dates_groups = orders[['date', 'group']].drop_duplicates()

# создадим кумулятивные данные по дням о заказах
orders_agg = (dates_groups
    .apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]
    .agg({'date': 'max', 'group': 'max', 'transactionId': 'nunique', 'visitorId': 'nunique', 'revenue': 'sum'}), axis=1)
    .sort_values(by=['date', 'group'])
)

# создадим кумулятивные данные по дням о посетителях
visitors_agg = (dates_groups
    .apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])]
    .agg({'date': 'max', 'group': 'max', 'visitors': 'sum'}), axis=1)
    .sort_values(by=['date', 'group'])
)

# объединим кумулятивные данные в одной таблице
cumulative_data = orders_agg.merge(visitors_agg, on=['date', 'group'])
# переименуем столбцы
cumulative_data.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

# посмотрим на получившийся датафрейм
display(cumulative_data.head())
```

### Посчитайте статистическую значимость различий в среднем количестве заказов между группами по «сырым» данным. Сделайте выводы и предположения.

Посчитаем статистическую значимость различий в среднем количестве заказов между группами по «сырым» данным с помощью теста Манна-Уитни.


Также сформулируем гипотезы. 
   - Н0: Различий в среднем количестве заказов между группами А и В нет.
   - H1: Есть различия в среднем количестве заказов между группами А и В

```python
# создадим датафреймы с числом визитов по дням отдельно по группам А и В
visitors_A_daily = visitors[visitors['group']=='A'][['date', 'visitors']]
visitors_A_daily.columns = ['date', 'visitors_per_date_A']

visitors_B_daily = visitors[visitors['group']=='B'][['date', 'visitors']]
visitors_B_daily.columns = ['date', 'visitors_per_date_B']
```

```python
# создадим датафреймы с числом заказов по каждому пользователю отдельно по группам А и В
orders_by_users_A = (orders[orders['group']=='A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': 'nunique'})
)
orders_by_users_A.columns = ['visitorId', 'orders']

orders_by_users_B = (orders[orders['group']=='B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': 'nunique'})
)
orders_by_users_B.columns = ['visitorId', 'orders']
```

```python
# создадим списки из числа заказов определенного пользователя (в том числе где 0 заказов) отдельно по группам А и В
sample_A = (pd.concat([orders_by_users_A['orders'],
       pd.Series(0, index=np.arange(visitors_A_daily['visitors_per_date_A'].sum() - len(orders_by_users_A['orders'])), 
       name='orders')],
       axis=0)
)

sample_B = (pd.concat([orders_by_users_B['orders'],
       pd.Series(0, index=np.arange(visitors_B_daily['visitors_per_date_B'].sum() - len(orders_by_users_B['orders'])),
       name='orders')],
       axis=0)
)
```

```python
# определим статистическую значимость различий в среднем количестве заказов между группами с помощью теста Манна-Уитни
print('p-значение теста Манна-Уитни:', "{0:.3f}".format(stats.mannwhitneyu(sample_A, sample_B)[1]))
```

```python
# определим относительное изменение среднего количества заказов
print('Относительное изменение среднего количества заказов группы B к группе A:', 
      "{0:.3f}".format(sample_B.mean() / sample_A.mean() - 1))
```

**Вывод:**

При критическом уровне статистической значимости alpha = 0.05, р-value меньше alpha. Мы отвергаем нулевую гипотезу "Различий в среднем количестве заказов между группами А и В нет." И принимаем альтернативную гипотезу "Есть различия в среднем количестве заказов между группами А и В".

Можно сделать вывод о различии между группами А и В. 

При этом Относительное изменение среднего количества заказов группы B к группе A больше на 13.8%.

```python
# соединим вместе выбросы
abnormal_users = (pd.concat([users_with_many_orders, users_with_expensive_orders], axis=0)
    .drop_duplicates()
    .sort_values()
)
```

```python
print('Количество аномальных пользователей:', len(abnormal_users))
print('Общее количество пользователей:', orders['visitorId'].nunique())
print('Доля аномальных пользователей к общему количеству пользователей в %:', 
      len(abnormal_users) / orders['visitorId'].nunique() * 100 )
```

Итак, в число аномальных пользователей по количеству заказов и по сумме заказа вошло 5% всех пользователей.

Теперь исключим эти пользователей из данных и узнаем как аномальные пользователи повлияли на результаты теста.

```python
# создадим списки из числа заказов определенного пользователя (в том числе где 0 заказов) отдельно по группам А и В
sample_A_filtered = (pd.concat([orders_by_users_A[np.logical_not(orders_by_users_A['orders'].isin(abnormal_users))]['orders'],
       pd.Series(0, index=np.arange(visitors_A_daily['visitors_per_date_A'].sum() - len(orders_by_users_A['orders'])), 
       name='orders')],
       axis=0)
)

sample_B_filtered = (pd.concat([orders_by_users_B[np.logical_not(orders_by_users_B['orders'].isin(abnormal_users))]['orders'],
       pd.Series(0, index=np.arange(visitors_B_daily['visitors_per_date_B'].sum() - len(orders_by_users_B['orders'])), 
       name='orders')],
       axis=0)
)
```

Теперь применим статистический критерий Манна-Уитни к очищенным от выбросов данным.

```python
# определим статистическую значимость различий в среднем количестве заказов между группами с помощью теста Манна-Уитни
# по очищенным данным 
print('p-значение теста Манна-Уитни по очищенным данным:',
      "{0:.3f}".format(stats.mannwhitneyu(sample_A_filtered, sample_B_filtered)[1]))
```

```python
# определим относительное изменение среднего количества заказов по очищенным данным 
print('Относительное изменение среднего количества заказов группы B к группе A по очищенным данным:', 
      "{0:.3f}".format(sample_B_filtered.mean() / sample_A_filtered.mean() - 1))
```

**Вывод:**

После очистки данных от выбросов результат теста не поменялся. 

При критическом уровне статистической значимости alpha = 0.05, р-value меньше alpha. Мы отвергаем нулевую гипотезу "Различий в среднем количестве заказов между группами А и В нет." И принимаем альтернативную гипотезу "Есть различия в среднем количестве заказов между группами А и В".

Можно сделать вывод о различии между группами А и В. При чем p-значение теста Манна-Уитни = 0.017 и по сырым данным, и по очищенным.

Относительное изменение среднего количества заказов группы B к группе A больше на 13.8% и осталось таким как и по сырым данным (13.8%).
